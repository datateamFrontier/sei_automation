```python
import pandas as pd
import numpy as np
import datetime as dt
import os
import olah_data as od

pd.set_option("display.max_columns", None)
```

```python
dir = "data kuesioner/"
os.listdir(dir)
```

```python
kategori = 'call center'
# kategori = 'twitter'
# kategori = 'email'

kategori_call_center = ['car insurance','regular banking','courier',
                       'life & helath insurance'] # benahi nama di kuesionernya untuk lhi
```

```python
data_raw = pd.read_csv("data kuesioner/"+"Penilaian "+kategori.title()+" Service Excelence Index.csv")
data_raw['Timestamp'] = data_raw['Timestamp'].str.replace(" MDT", "")
data_raw['Timestamp'] = pd.to_datetime(data_raw['Timestamp']) + pd.Timedelta('15 hour')
data_raw.insert(loc=1, column="Waktu", value = [d.time() for d in data_raw['Timestamp']])
data_raw.insert(loc=2, column="Tanggal", value = [d.date() for d in data_raw['Timestamp']])
data_raw.insert(loc=3, column="Bulan", value = pd.DatetimeIndex(data_raw['Tanggal']).month)
data_raw = data_raw.drop(labels='Timestamp', axis=1)
```

# call center

```python
kolom_ivr = od.kolom_ivr
```

```python
data_cc = data_raw[data_raw['Kategori Call Center'].isin([kategori_call_center[0].title()])].copy()

call_center_lain = ["Call Center - "+x.title() for x in kategori_call_center if x!=kategori_call_center[0]]
data_cc = data_cc.drop(labels=['Nama Caller','Kategori Call Center']+call_center_lain, axis=1)
data_cc = data_cc.rename(columns={"Call Center - "+kategori_call_center[0].title():"Brand"})
```

```python
brand_unik = data_cc['Brand'].unique()
```

```python
bobot = pd.read_excel("bobot/bobot perhitungan "+kategori+".xlsx")
bobot = bobot[['NOMOR LEVEL 1','NOMOR LEVEL 2','KODE','ASPEK','SUB ASPEK','RINCIAN','SUB RINCIAN']]
bobot = bobot.fillna(method="ffill")
# bobot.head()

aspek = bobot['ASPEK']
jawaban = bobot[['SUB ASPEK','RINCIAN','SUB RINCIAN']]

new_kol = bobot['KODE']
```

```python
kolom_raw = data_cc.columns[4:].to_list()
for brd in brand_unik:
    df_satu_brand = data_cc[data_cc["Brand"].isin([brd])]
    df_satu_brand = df_satu_brand.sort_values(by=['Tanggal','Waktu'])
#     display(df_satu_brand)
    for klm_raw in kolom_raw:
        pertanyaan = klm_raw.split(". ")[1]
        if pertanyaan not in aspek.unique():
            print(pertanyaan)
```
